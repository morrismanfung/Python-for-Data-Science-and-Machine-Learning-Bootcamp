```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style( 'whitegrid')
```

```python
titanic = sns.load_dataset( 'titanic')
```

```python
titanic.head()
```

### Below is a joint plot showing the relationship between age and fare.

```python
sns.jointplot( x = 'fare', y = 'age', data = titanic)
plt.title( 'Relationship between age and fare')
```

### Below is a distribution plot of the fare.

```python
sns.displot( titanic[ 'fare'], bins = 30, kde = True)
plt.title( 'Distribution of fare')
```

### Below is a boxplot showing distribution of age amond each class.
It can show that class and age could be correlated, with older people tend to be in the first class.

```python
sns.boxplot( x = 'class', y = 'age', data = titanic, palette = 'rainbow')
plt.title( 'Age and Classes')
```

### Below is a swarm showing distribution of age amond each class.
It can show that class and age could be correlated, with older people tend to be in the first class.

```python
sns.swarmplot( x = 'class', y = 'age', data = titanic, palette = 'Set2', size = 3)
plt.title( 'Age and Classes')
```

### Below is a bar plot of count of each sex.
It can be concluded that there were more male participants.

```python
sns.countplot( titanic['sex'])

plt.title( 'Count of male and female')
```

### Below is a visualization of the correlation matrix.
Please note that only the quantitative variables are presented.

```python
sns.heatmap( titanic.corr(), cmap = 'coolwarm')
plt.title( 'titanic.corr()')
```

### Below are 2 ways to visualize the distribution of age in male and female separately.

```python
g = sns.FacetGrid( data = titanic, col = 'sex')
g.map( sns.distplot, 'age', kde = False, bins = 15)
```

```python
g = sns.FacetGrid( data = titanic, col = 'sex')
g.map( plt.hist, 'age', bins = 15)
```
