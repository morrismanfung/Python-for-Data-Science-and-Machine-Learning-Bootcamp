# Visualization Exercise with Seaborn

In this exercise, I will use the python library *seaborn* to visualize a data set *titanic*. It is a classic data set for studying machine learning. The data set can be found on https://www.kaggle.com/c/titanic.

# Basic set-ups

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
# The data can be directly loaded as it is stored in the seaborn libary
```

# Plots

```python
sns.jointplot( x = 'fare', y = 'age', data = titanic)

plt.subplots_adjust(top=0.9)
plt.suptitle('Joint plot between Fare and Age', fontsize = 16)
```

```python
sns.distplot( titanic[ 'fare'], bins = 30, color = 'red', kde = False)
plt.title( 'Fare', fontsize = 16)
```

```python
sns.boxplot( x = 'class', y = 'age', data = titanic, palette = 'rainbow')
plt.title( 'Box plot of Classes', fontsize = 16)
```

```python
sns.swarmplot( x = 'class', y = 'age', data = titanic, palette = 'Set2')
plt.title( 'Swarm plot of Classes', fontsize = 16)
```

```python
sns.countplot( 'sex', data = titanic)
plt.title( 'Distribution of Gender', fontsize = 16)
plt.xlabel( 'Gender')
```
